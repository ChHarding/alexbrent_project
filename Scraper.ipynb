```python
import requests as r
from bs4 import BeautifulSoup
import pandas as pd
```

```python
data= []
for i in range(1,51):
    print(i)
    res = r.get(f"https://books.toscrape.com/catalogue/page-{i}.html").text
    soup = BeautifulSoup(res)

    book_container = soup.find_all("article",class_ ="product_pod")
    for book in book_container:
        img_url = "https://books.toscrape.com/"+book.find("img",class_="thumbnail")['src']
        book_title = book.find("h3").text
        data.append(
        {    "img_url":img_url,
            "book_title":book_title}
        )
                
    print(f"Scraped {i}/50 pages")
    
    
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('./library_db.csv',index=False)
```
